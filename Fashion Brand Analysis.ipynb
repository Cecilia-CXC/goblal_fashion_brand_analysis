```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the excel file
df = pd.read_excel('Dataset Global Fashion Brands Brand Equity Ranking Growth Rate  COO ROO 2001-2021.xlsx')
```

```python
#re-shape into a long and narrow dataset
# Identify static columns (non-year)
id_vars = ['BrandName', 'BrandOriginCountry', 'BrandOriginRegion', 'BrandSector', 'BrandSubSector']

# Extract year-based columns
rank_cols = [c for c in df.columns if c.startswith('Rank')]
equity_cols = [c for c in df.columns if c.startswith('Equity')]
growth_cols = [c for c in df.columns if c.startswith('GrowthRate')]

# Create a tidy dataframe by looping through years
records = []

for year in range(2001, 2022):
    data = df[id_vars].copy()
    data["Year"] = year
    data["Rank"] = df.get(f"Rank{year}")
    data["Equity"] = df.get(f"Equity{year}")
    data["GrowthRate"] = df.get(f"GrowthRate{year}")
    records.append(data)

# Combine all years into one dataframe
tidy_df = pd.concat(records, ignore_index=True)

# Optional: sort by brand and year
tidy_df = tidy_df.sort_values(by=["BrandName", "Year"]).reset_index(drop=True)
```

```python
#Imputation for missing values on Rank column using forward fill
tidy_df['Rank'].fillna(method='ffill', inplace=True)
```

```python
#Use Multivariate Imputation with Chained Equations (MICE) to impute Equity & GrowthRate
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp = IterativeImputer(random_state=0)
tidy_df[['Equity', 'GrowthRate']] = imp.fit_transform(tidy_df[['Equity', 'GrowthRate']])
```
